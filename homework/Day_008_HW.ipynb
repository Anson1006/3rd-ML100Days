```python
data=df['AMT_INCOME_TOTAL']
#data.max()
log_data=np.log(data)
plt.hist(log_data,alpha=0.5,histtype='bar',color='steelblue',edgecolor='none',)
plt.xlabel('amount',fontsize=15)
plt.ylabel('log_income',fontsize=15)
plt.title("histogram of AMT_INCOME_TOTAL ",fontsize=18) 
plt.show()
```
